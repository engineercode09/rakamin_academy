# Business Insight

```python
fig = plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x='Revenue')
for i in ax.patches:
  ax.text(x=i.get_x()+i.get_width()/2,
          y=i.get_height()/2,
          s=f"{np.round(i.get_height()/len(df)*100,2)}%",
          ha="center",
          va="center",
          size=20,
          weight="bold",
          rotation="360",
          color="white")
plt.title("Revenue Feature", size=20)
plt.show()
```

Analisa:
* Data target (Revenue) terdiri dari 84,53% nilai False (customer yang tidak melakukan pembelian) dan 15,47% nilai True (customer yang melakukan pembelian)

```python
def ratio_revenue(data, count, by):
  result = []
  for i in range(len(data)):
    result.append(data[count][i] / data[data[by] == data[by][i]][count].sum() * 100)
  return result
```

```python
df_v = df.groupby(['Month','Revenue']).aggregate({'Revenue':'count'})
df_v.rename(columns={'Revenue':'Total'},inplace=True)
df_v.reset_index(inplace=True)
df_v['Ratio'] = ratio_revenue(df_v, 'Total', 'Month')
```

```python
plt.figure(figsize=(8,20))
sns.barplot(x='Ratio', y='Month', hue='Revenue', data=df_v)
for i in range(0, len(df_v), 2):
  plt.text(df_v.Ratio[i]+0.5, (i/2)-0.2, '{:,.2f}%'.format(df_v.Ratio[i]))
  plt.text(df_v.Ratio[i+1]+0.5, (i/2)+0.2, '{:,.2f}%'.format(df_v.Ratio[i+1]))
plt.xlim([0,150])
plt.title('Persentase Revenue Berdasarkan Month', size=20)
plt.xlabel('Percentage')
plt.show()
```

Analisa:
* Pada bulan Jul, Aug, Sept, Oct, Nov jumlah revenue lebih dari 15% Visitor

```python
df_v = df.groupby(['VisitorType','Revenue']).aggregate({'Revenue':'count'})
df_v.rename(columns={'Revenue':'Total'},inplace=True)
df_v.reset_index(inplace=True)
df_v['Ratio'] = ratio_revenue(df_v, 'Total', 'VisitorType')
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Ratio', y='VisitorType', hue='Revenue', data=df_v)
for i in range(0, len(df_v), 2):
  plt.text(df_v.Ratio[i]+0.5, (i/2)-0.2, '{:,.2f}%'.format(df_v.Ratio[i]))
  plt.text(df_v.Ratio[i+1]+0.5, (i/2)+0.2, '{:,.2f}%'.format(df_v.Ratio[i+1]))
plt.xlim([0,120])
plt.title('Persentase Revenue Berdasarkan VisitorType', size=20)
plt.xlabel('Percentage')
plt.show()
```

Analisa:
* New_Visitor / pengunjung baru lebih cendrung membeli / menghasilkan revenue dibandingkan tipe pengunjung yang lain

```python
df_v = df.groupby(['Weekend','Revenue']).aggregate({'Revenue':'count'})
df_v.rename(columns={'Revenue':'Total'},inplace=True)
df_v.reset_index(inplace=True)
df_v['Ratio'] = ratio_revenue(df_v, 'Total', 'Weekend')
df_v['Weekend'] = df_v['Weekend'].apply(lambda x: 'True' if x else 'False')
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='Ratio', y='Weekend', hue='Revenue', data=df_v)
for i in range(0, len(df_v), 2):
  plt.text(df_v.Ratio[i]+0.5, (i/2)-0.2, '{:,.2f}%'.format(df_v.Ratio[i]))
  plt.text(df_v.Ratio[i+1]+0.5, (i/2)+0.2, '{:,.2f}%'.format(df_v.Ratio[i+1]))
plt.xlim([0,120])
plt.title('Persentase Revenue Berdasarkan Weekend', size=20)
plt.xlabel('Percentage')
plt.show()
```


Analisa:
* Tidak ada perbedaan yang signifikan dari jumlah persentase reveneu saat weekend atau bukan

Recommendations:
1. Lebih baik dilakukan suatu marketing khusus pada bulan-bulan yang memiliki persentase revenue yang rendah
2. Perlu dilakukan marketing khusus kepada returning visitor karena dari hasil analisa, visitor sangat banyak yang bertipe returning namun persentanse revenuenya paling rendah
3. Perlu dibuatkan strategi marketing pada hari-hari non-weekend karena dari hasil analisa, jumlah visitor pada saat non-weekend sangat banyak dan berbeda jauh dengan jumlah visitor pada saat weekend namun persentase revenue pada saat weekend justru sedikit lebih besar dengan saat non-weekend
